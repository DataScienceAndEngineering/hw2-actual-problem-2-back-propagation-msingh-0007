## a. Proof of Michael Nielsons equation BP3


Deriving BP3 is very similar to how BP1 was derived, except we also do partial derivative on Z with respect to bias b since Z also depends on b


$ \frac{\partial C}
{\partial b_j^L}= \Sigma_k \frac{\partial C}
{\partial a_k^L}*\frac{\partial a_k^L}
{\partial z_j^L}*\frac{\partial z_j^L}
{\partial b_j^L} $

$z_j^L = \Sigma_k w_{jk}^L * a_k^{L-1} + b_j^L $

$\frac{\partial z_j^L}{\partial b_j^L} = 1$

Thus:

$ \frac{\partial C}
{\partial b_j^L}= \Sigma_k \frac{\partial C}
{\partial a_k^L}*\frac{\partial a_k^L}
{\partial z_j^L} = \delta_j^L $

To solve this lets seperate $\frac{\partial C}
{\partial \omega_{jk}^L}$ using chain rule:

$\frac{\partial C}
{\partial \omega_{jk}^L} = \Sigma_k \frac{\partial C}
{\partial a_k^L}*\frac{\partial a_k^L}

$z_j^L = \Sigma_k w_{jk}^L * a_k^{L-1} + b_j^L $

$\frac{\partial z_j^L}{\partial w_{jk}^L} = a_k^{L-1}$

$\delta_j^L = \frac{\partial C}

Now we will modify the backpropagation code using the matrix approach to merge the weights and biases. We will start the Weights and Bias matrix as a randomized 5 x 3 matrix ( 4 Inputs + 1 Bias -> 3 Activation outputs)

The Third Layer will have a Weight-Bias Matrix of 4 x 3 (3 Activation Outputs from Layer 2 + 1 Bias Term -> 3 Activation Outputs for Layer 3)

This would also mean that we would have to Augment the activation matrix of Layer 2 with another row of 1s for the biases

```python
def backprop(x,y):

        # feedforward
        np.random.seed(42)
        W_2 = np.matrix([np.random.rand(5) , np.random.rand(5), np.random.rand(5)]) # w_2: 5x3 since 4 inputs + 1 bias to 3 activations in Layer 2
        z_2 = np.dot(W_2, x) # z = Input-Bias Matrix . Weight Vector
        a_2 = sigmoid(z_2)

        a_2_bias = np.row_stack((a_2, np.ones(150))) #Augmenting the Activation matrix with Bias Input = 1
        W_3 = np.matrix([np.random.rand(4) , np.random.rand(4), np.random.rand(4)]) # w_3: 4x3 since 3 activations + 1 bias to 3 outputs in Layer 3
        z_3 = np.dot(W_3, a_2_bias)
        a_3 = sigmoid(z_3)
        print("FeedForward Successful")
        # backward pass
        delta_3 = np.multiply(cost_derivative(a_3, y),sigmoid_prime(z_3)) #Error matrix for Layer_3 3x150
        delta_2 = np.multiply(np.dot(np.transpose(W_3),delta_3), np.row_stack((sigmoid_prime(z_2), np.zeros(150)))) #Error matrix for Layer_2
        #also have to augment sigmoid_prime data due to biases, since the bias "activation" input = 1 and is a constant, we can augment a row of zeros
        #since the derivative of a constant term is 0
        gradient_3 = np.dot(a_2_bias,np.transpose(delta_3))
        gradient_2 = np.dot(x, np.transpose(delta_2))
        return (gradient_3, gradient_2)



def cost_derivative(output_activations, y):
        """Return the vector of partial derivatives \partial C_x /
        \partial a for the output activations."""

        return (output_activations-np.transpose(y))

def sigmoid(z):
    """The sigmoid function."""
    return 1.0/(1.0+np.exp(-z))

def sigmoid_prime(z):
    """Derivative of the sigmoid function."""
    return np.multiply(sigmoid(z),(1-sigmoid(z)))
```

We now output the matrices for the Gradients of all Weights and Biases. The Layer 3 Gradients will be a 4 x 3 Matrix while the Layer 2 Gradients will be a 5 x 3 Matrix

```python
dC_dW3, dC_dW2 = backprop(Input_Matrix, Output_Matrix)
print(dC_dW3) # Each Row are Gradients to applied to weights from 1 input, the last row depicts the Gradients for the Biases
print(' ')
print(dC_dW2)
```
